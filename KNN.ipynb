```python
from math import sqrt
# calculate the Euclidean distance between two vectors
def euclidean_distance(row1, row2):
  distance = 0.0
  for i in range(len(row1)-1):
    distance += (row1[i] - row2[i])**2
  return sqrt(distance)

def cb_distance(row1, row2):
  return sum(abs(row1[i] - row2[i]) for i in range(len(row1)-1))
```

```python
# Locate the most similar neighbors
def get_neighbors(train, test_row, num_neighbors):
	distances = list()
	for train_row in train:
		dist = euclidean_distance(test_row, train_row)
		distances.append((train_row, dist))
	distances.sort(key=lambda tup: tup[1])
	neighbors = list()
	for i in range(num_neighbors):
		neighbors.append(distances[i][0])
	return neighbors
```

```python
# Make a classification prediction with neighbors
def predict_classification(train, test_row, num_neighbors):
	neighbors = get_neighbors(train, test_row, num_neighbors)
	output_values = [row[-1] for row in neighbors]
	prediction = max(set(output_values), key=output_values.count)
	return prediction

def predictions(train, test, num_neighbors):
	return [predict_classification(train, test_row, num_neighbors) for test_row in test]
```

```python
import pandas as pd

abalone=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header = None)
print(abalone[4])
# c = Counter( abalone[2] )

# print( c.items() )
for i in range(len(abalone[4])):
    if abalone[4][i] == 'Iris-setosa':
        abalone[4][i] = 1
    elif abalone[4][i] == 'Iris-versicolor':
        abalone[4][i] = 2
    elif abalone[4][i] == 'Iris-virginica':
        abalone[4][i] = 3
        
abalone[4]
# abalone.head()
```

```python
X=abalone.iloc[:,:-1]
Y=abalone.iloc[:,4]
Y=Y.astype('int')
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
X_train["Y"]=Y_train
X_test["Y"]=np.zeros((len(Y_test)),np.int16)
Y_train=Y_train.astype(np.int16)
Y_test=Y_test.to_list()
prediction_custom=predictions(X_train.to_numpy(), X_test.to_numpy(), num_neighbors=9)
for i in range(len(Y_test)):
    print(Y_test[i], prediction_custom[i])
custom_score_iris = accuracy_score(Y_test,prediction_custom)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=9)
from sklearn.metrics import accuracy_score
neigh.fit(X_train,Y_train)
prediction=neigh.predict(X_test)
print(Y_test)
print((prediction))
sklearn_score_iris = accuracy_score(Y_test,prediction)
```

```python
for i in range(len(prediction)):
    print(prediction_custom[i]," ",prediction[i])
```

```python
print("accuracy of custom developed code is {}".format(custom_score_iris))
print("Accuracy of sklearn score is {}".format(sklearn_score_iris))
```
